```python
import numpy as np
vi = np.array([1,0,0.1]) # initial velocity
B = np.array([0,0,1]) # magnetic field
q = -1 # free code camp to numpy
m = 1
E = np.array([0,0,0.01])
ti = 0
tf = 100
t = ti
v = vi
delta_t = 0.001
vel_save = [vi]

# update rule
def update(vk, delta_t):
  return vk+q/m*(E+np.cross(vk,B))*delta_t # you can also calculate by hand

# loop start
while (t<tf):
  v = update(v,delta_t)
  vel_save.append(v)
  t = t + delta_t
print(v)
print(vel_save)
```

```python
xi = np.array([0,0,0])
x_save = [xi]
# loop start
for velocity in vel_save:
  x_save.append(x_save[-1]+velocity*delta_t)
#x_save
```
